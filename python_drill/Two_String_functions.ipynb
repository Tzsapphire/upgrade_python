Exercise 10

Create this string (with spaces):

Remove spaces from the left

Remove spaces from the right

Remove spaces from both sides

Print each result.

```python
space = " python  "
print (len(space))
print(space.lstrip())
print (len((space.lstrip())))
print(space.rstrip())
print (len((space.rstrip())))
print(space.strip())
print (len((space.strip())))
```
