````python
sales = [120, 450, 800, 50, 900, 300]
low =[]
medium = []
high = []
for price in sales:
    if price < 300:
        low.append(price)
        print(price, "is low")
    elif 300 <= price <= 700:
        medium.append(price)
        print(price, "is Medium")
    else:
        high.append(price)
        print(price, "is High")
print(f'Low prices are :{low}, \n Medium prices are :{medium}, \n High prices are :{high}')
````

````python
numbers = [12, 7, 9, 20, 33, 14, 5]
even_numbers = []
for number in numbers:
    if number % 2 ==0:
        even_numbers.append(number)
        print(number)
print('The list of even numbers are: ', even_numbers.sort())
print('The list of even numbers are: ', sorted(even_numbers))
print('The list of even numbers are: ', sorted(even_numbers, reverse = True))
````
