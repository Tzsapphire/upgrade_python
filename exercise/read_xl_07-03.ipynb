```python
def extract_xls():
    filename = input("file to extract is: ")
    excel_file =  ('./excel_files/' + filename + '.xlsx')
    file = pd.read_excel(excel_file)
    # print(file.head(3), '\n')
    # print(file.info(), '\n')
    return file
```

```python
def transform_excel(file, treatment_type, outcome):
    treatment_outcome = file[(file['Treatment Type']== treatment_type) & (file['Outcome'] == outcome)]
    print(treatment_outcome.head())
    return treatment_outcome
```

```python
def load_excel(file):
    names = input("whats your filename: ")
    # save = file.to_csv()
    # print(save.head())
    # csv_f = file.to_csv(names + '.csv')
    file.to_excel(names + '.xlsx')
    return f"file is saved as {names}.xlsx"
    
```
