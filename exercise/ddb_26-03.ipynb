```python
new_db
db_conn = f"mssql+pymssql://sa:Password123!@localhost/{new_db}"
db_engine = create_engine(db_conn)
```

```python
db_engine.connect()
```

```python
new_schema = 'first'

sql = f"""
IF NOT EXISTS (
    SELECT 1
    FROM sys.schemas
    WHERE name = '{new_schema}'
)
EXEC('create schema [{new_schema}]')
"""
try:
    with db_engine.connect().execution_options(isolation_level="AUTOCOMMIT") as conn:
        conn.exec_driver_sql(sql)
except Exception as e:
    print(e)
```
