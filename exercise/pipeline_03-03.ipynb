## Pandas tutorial
Here I learn how to use pandas to perform an ETL. 

In addition, I create a function that can help perform this ETL.

```python
import pandas as pd
```

```python
file = pd.read_csv('customers.csv')
file.head()
```

```python
file.columns
```

```python
customer_info = file[[ 'Index', 'First Name', 'Last Name', 'Country','Phone 1', 'Email', 'Subscription Date']]
```

```python
customer_info.head()
```

```python
newly_subscribed = customer_info[customer_info["Subscription Date"] > '2022-01-01']
```

```python
newly_subscribed.head()
```

```python
newly_subscribed.count()
```

```python
newly_subscribed.to_csv('new_subscribers_2022.csv', index = False)
```

Function for the ETL pipeline:

```python
def extract_data(csv_file):
    file = pd.read_csv(csv_file)
    print(file.head(3))
    return "file reading successful"
```

```python
extract_data('customers.csv')
```

```python
def transform(file):
    print("getting newly subscribed customers since 2022")
    newly_subscribed = file[file["Subscription Date"] > '2022-01-01']
    print(newly_subscribed.head(3))
    return "transformation successful"
```

```python
transform(file)
```

```python
def load_data(newly_subscribed):
    file_name = input("whats your output file name?: ")
    if '/' in file_name or '\\' in file_name:
        print("Invalid characters like / or \ in file_name")
    else:    
        newly_subscribed.to_csv(f'{file_name}.csv', index = False)
    return f"file successfully stored as {file_name}.csv"
```

```python
load_data(newly_subscribed)
```
