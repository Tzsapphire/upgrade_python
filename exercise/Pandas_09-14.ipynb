```python
chicago['people'] = "government"
chicago.head(3)
```

```python
chicago.drop(columns='people' , inplace=True)
chicago.head(3)
```

```python
# to copy a df into another variable use the .copy()
chicago_new = chicago.copy()
```

```python
chicago_new['first_name'] = chicago_new['Name'].str.split(', ').str[1]
chicago_new['last_name'] = chicago_new['Name'].str.split(', ').str[0]
```

```python
chicago_new.drop(columns='last_name', inplace = True)
chicago_new.head(3)
```

```python
# on my new copied df i am adding a new column
chicago_new['sal_name'] = chicago_new['first_name'] + ' = ' + chicago_new['Department']
```

```python
chicago_new.head()
```

```python
chicago_new.rename(columns = {'sal_name' : 'dept_name'}, inplace = True)
```
