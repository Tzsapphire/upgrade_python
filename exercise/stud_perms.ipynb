```python
def gender_based_score():
    perms = pd.read_csv('Students Performance.csv')
    perms.columns = perms.columns.str.replace(' ', '_')
    perms.columns = perms.columns.str.lower()
    perms.rename({'race/ethnicity': 'ethnicity'}, axis='columns', inplace=True )
    scores =['math_score', 'reading_score', 'writing_score']
    perms['total_score'] = perms[scores].sum(axis=1)
    perms['average_score'] = perms[scores].mean(axis=1)
    scores1 = ['math_score','reading_score', 'writing_score', 'total_score', 'average_score']
    scores_by_gender = perms.groupby(['gender'])[scores1].sum()
    scores_by_gender.to_csv('scores_by_gender')
    
    return True 
```
